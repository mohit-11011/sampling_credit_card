UNDERSAMPLING

```python
class_count_0, class_count_1 = dataset['Class'].value_counts()
class_0 = dataset[dataset['Class'] == 0]
class_1 = dataset[dataset['Class'] == 1]# print the shape of the class
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

```python
class_0_under = class_0.sample(class_count_1)

test_under = pd.concat([class_0_under, class_1], axis=0)

print("total class of 1 and0:",test_under['Class'].value_counts())
test_under
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn import svm
from sklearn.naive_bayes import BernoulliNB
def accuracy_for_models(x_dataset,y_dataset):
    accuracy=[]
    X_train, X_test, y_train, y_test = train_test_split(x_dataset, y_dataset, random_state=6, train_size = .6)

    xgb_model = XGBClassifier().fit(X_train, y_train)
    xgb_y_predict = xgb_model.predict(X_test)
    xgb_score = accuracy_score(xgb_y_predict, y_test)
    accuracy.append(xgb_score)

    rfc = RandomForestClassifier()
    rfc.fit(X_train, y_train)
    rfc_y=rfc.predict(X_test)
    rfc_score=accuracy_score(rfc_y,y_test)
    accuracy.append(rfc_score)

    clf = DecisionTreeClassifier()
    clf.fit(X_train,y_train)
    clf_y= clf.predict(X_test)
    clf_score=accuracy_score(clf_y,y_test)
    accuracy.append(clf_score)

    svmi = svm.SVC(kernel='linear', C=1, random_state=42)
    svmi.fit(X_train, y_train)
    svm_pred = svmi.predict(X_test)
    svm_acc=accuracy_score(svm_pred,y_test)
    accuracy.append(svm_acc)

    classifier = BernoulliNB()
    classifier.fit(X_train, y_train)
    bayesian_pred = classifier.predict(X_test)
    bayesian_acc = accuracy_score(bayesian_pred,y_test)
    accuracy.append(bayesian_acc)    

    return accuracy




    

```

```python
y_under=test_under["Class"]
test_under.pop("Class")
x_under=test_under
under_sampling=accuracy_for_models(x_under,y_under)
under_sampling
```

OVERSAMPLING

```python
class_1_over = class_1.sample(class_count_0, replace=True)
test_over = pd.concat([class_1_over, class_0], axis=0)
test_over
```

```python
y_over=test_over["Class"]
test_over.pop("Class")
x_over=test_over
over_sampling=accuracy_for_models(x_over,y_over)
over_sampling
```

TOMEK LINKS

```python
from imblearn.under_sampling import TomekLinks
tl=TomekLinks()
x_tomek,y_tomek = tl.fit_resample(x_dataset,y_dataset)
tomek_acc=accuracy_for_models(x_tomek,y_tomek)
print(y_tomek.size)
tomek_acc
```

SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(x_dataset, y_dataset)
print(y_smote.size)
smote_acc=accuracy_for_models(x_smote,y_smote)
smote_acc
```

NEAR MISS

```python
from imblearn.under_sampling import NearMiss

nm = NearMiss()

x_nearmiss, y_nearmiss = nm.fit_resample(x_dataset, y_dataset)
print(y_nearmiss.size)
nearmiss_acc=accuracy_for_models(x_nearmiss,y_nearmiss)
nearmiss_acc
```
